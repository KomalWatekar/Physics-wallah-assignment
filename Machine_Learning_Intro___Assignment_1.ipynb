1.Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS)

1. Artificial Intelligence (AI)
The Broad Goal: AI is the "umbrella" term. It refers to the broad concept of creating machines or software capable of mimicking human intelligence to perform tasks.


Core Idea: Making a machine "smart."

Methods: It can include everything from simple "if-then" rules (symbolic AI) to complex biological simulations.

Example: A chess-playing program or a voice assistant like Siri.

2. Machine Learning (ML)
The Learning Method: ML is a subset of AI. Instead of programming a computer with every possible rule, you feed it data and allow it to "learn" the rules through statistical patterns.


Core Idea: Learning from data to make predictions or decisions without being explicitly programmed for that specific task.

Methods: Linear Regression, Decision Trees, and Support Vector Machines.

Example: Netflix suggesting a movie based on your watch history or a Gmail spam filter.

3. Deep Learning (DL)
The Advanced Subset: DL is a subset of ML. It uses multi-layered "Artificial Neural Networks" inspired by the structure of the human brain to process data in a more complex way.


Core Idea: "Deep" refers to the many layers of neurons that help the machine learn extremely complex patterns (like the nuances of a human face).

Methods: Convolutional Neural Networks (CNNs) for images, Transformers for text (like ChatGPT).

Example: Self-driving cars identifying pedestrians or real-time language translation.

4. Data Science (DS)
The Practical Application: Data Science is an interdisciplinary field that uses AI, ML, and DL—along with statistics and domain expertise—to extract insights and value from data.

Core Idea: Using data to solve business problems or answer "What if?" questions.

Scope: It covers the entire lifecycle of data: collection, cleaning, analysis, and visualization.

Example: A retail company analyzing sales data to decide where to open a new store.

Question 2: What are the types of machine learning? Describe each with one real-world example.
ans- Machine learning is typically divided into four main types based on how the algorithm learns and the type of data it uses.

1. Supervised Learning
In supervised learning, the algorithm is trained on labeled data. This means the "answers" are already provided in the training set, and the machine learns to map inputs to the correct outputs.


How it works: You show the machine thousands of pictures of cats and dogs, each labeled "Cat" or "Dog." Eventually, when you show it a new, unlabeled picture, it can identify the animal.

Real-World Example: Email Spam Filters. Your inbox has been trained on millions of emails labeled as "Spam" or "Not Spam." It looks for patterns (like certain keywords or suspicious senders) to automatically divert unwanted mail to your spam folder.


2. Unsupervised Learning
In unsupervised learning, the algorithm deals with unlabeled data. The machine is not given the "right answer"; instead, it tries to find hidden patterns, structures, or clusters within the data on its own.


How it works: You give the machine a massive list of customer purchase histories without any categories. The machine notices that one group buys mostly organic vegetables while another buys mostly snacks, effectively "grouping" them for you.

Real-World Example: Customer Segmentation in Marketing. E-commerce sites like Amazon use this to group shoppers by behavior. They might find a cluster of "budget travelers" vs. "luxury shoppers" without a human ever telling the machine what those categories are.



3. Semi-Supervised Learning
This is a hybrid approach. It uses a small amount of labeled data combined with a very large amount of unlabeled data. It’s useful when labeling data is too expensive or time-consuming for humans to do at scale.


How it works: A doctor might label a few hundred X-ray images as "healthy" or "pneumonia." The algorithm then uses those "known" examples to help it categorize the thousands of other unlabeled X-rays in the database.

Real-World Example: Facial Recognition. On platforms like Google Photos, you might label a few photos of a specific friend. The system then uses semi-supervised learning to find that same person across thousands of other unlabeled photos in your library.



4. Reinforcement Learning (RL)
Reinforcement learning is based on trial and error. An "agent" learns to make decisions in an environment to achieve a goal. It receives "rewards" for good actions and "penalties" for mistakes.



How it works: Think of it like training a dog with treats. If the dog sits (good action), it gets a treat (reward). If it ignores you, it gets nothing. Over time, the dog learns the best behavior to maximize treats.

Real-World Example: Self-Driving Cars. The car’s AI is "rewarded" for staying in the lane and reaching the destination safely, while it is "penalized" for hitting a curb or braking too late. Through millions of simulations, it learns the safest way to drive.

Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.
Ans- When I build a model, the goal is for it to perform well on new data. The main challenges I face in achieving this are:

Underfitting: This happens when my model is too simple to see the trend in the data. It has High Bias, meaning it makes wrong assumptions because it hasn't learned enough. It performs poorly on both training and test data.

Overfitting: This is the opposite problem. The model is too complex and "memorizes" the training data, including the noise and random fluctuations. It has High Variance. While it looks perfect on training data, it fails miserably when I show it new, unseen data.

The Bias-Variance Tradeoff is the balancing act I have to perform. If I try to lower the bias by making the model more complex, the variance usually goes up (leading to overfitting). If I try to lower the variance by simplifying the model, the bias goes up (leading to underfitting). My goal is to find the "sweet spot" where both errors are minimized so the model generalizes well to the real world.

Question 4: What are outliers in a dataset, and list three common techniques for handling them.
Ans- Understanding and Managing Outliers
In any dataset, I often encounter "the odd ones out"—data points that don't seem to fit the pattern established by the majority of the data. These are known as outliers.

What are Outliers?
An outlier is a data point that deviates significantly from the other observations in a dataset. They can occur for several reasons:

Data Entry Errors: Simple human mistakes, like typing "1000" instead of "100."

Measurement Errors: Faulty equipment or sensors providing incorrect readings.

Natural Variation: Sometimes a data point is a genuine extreme value, such as the income of a billionaire in a study of average salaries.
Why Handling Outliers is Critical
If I ignore outliers, they can seriously skew my results. For example, the Mean is highly sensitive to extreme values; a single outlier can make an entire dataset's average look much higher or lower than it actually is. This can lead to misleading conclusions and poor machine learning performance.

Three Common Techniques for Handling Outliers
When I identify outliers, I use these three techniques to manage them:

1. Trimming (Removal)
This is the simplest approach where I completely remove the outliers from the dataset.

When to use it: I only do this if I am certain the outlier is a mistake (like a data entry error) or if the dataset is large enough that losing a few points won't impact the overall statistical power.

Risk: If I remove genuine extreme values, I might lose important information about the diversity of the data.

2. Imputation (Replacement)
Instead of deleting the data, I treat the outlier like a missing value and replace it with a more representative number, such as the Median or the Mean of the rest of the dataset.

When to use it: This is useful when I want to keep the sample size consistent but need to neutralize the distorting effect of the extreme value.

Pro Tip: Using the Median is usually safer than the Mean because the Median itself is not affected by outliers.

3. Winsorization (Capping)
In this technique, instead of removing the outliers, I "cap" them at a certain percentile. For example, I might set any value above the 95th percentile to be equal to the 95th percentile value.

When to use it: This is ideal when the extreme values are valid but I want to limit their influence on the model. It keeps the data points in the set but brings them "back into line" with the rest of the distribution.

Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.
Ans- In real-world data science, I rarely receive a "perfect" dataset. Missing data is a common hurdle, and if I don't handle it properly, it can lead to biased models or even cause some algorithms to crash entirely.

The Process of Handling Missing Values
When I encounter missing data, I follow a systematic three-step process:

Identification: I first visualize the data (using tools like heatmaps) to see where the gaps are. I ask: Is the data missing randomly, or is there a pattern? (e.g., people with higher incomes might be less likely to report it).

Assessment: I evaluate the "percentage of missingness." If a column is missing 70% of its data, it is usually better to drop the entire feature. If only 1% is missing, it might be safe to simply remove those few rows.

Treatment: Depending on the context, I choose between Deletion (removing rows/columns) or Imputation (filling in the blanks).

Imputation Techniques
Instead of deleting data and losing valuable observations, I use imputation to estimate the missing values. Here are the techniques I use depending on the data type:

1. For Numerical Data: Median Imputation
For numerical features (like Age, Price, or Temperature), I often replace the missing values with the Median of that column.

Why Median? While I could use the Mean, the Median is much more robust. If there are outliers in the data, the Mean will be skewed, whereas the Median remains a reliable representation of the "middle" value.

Example: If I have a dataset of house prices and five entries are missing, I calculate the median price of all other houses and use that figure to fill the gaps.

2. For Categorical Data: Mode Imputation
For categorical features (like Color, Gender, or City), I use Mode Imputation, which involves filling the missing slots with the most frequently occurring category.

Why Mode? Since I cannot calculate an "average" for text or categories (you can't find the average of "Red" and "Blue"), the most logical choice is to use the category that appears most often in the dataset.

Example: In a dataset of customer locations where "Mumbai" is the most frequent city, I would fill any missing city entries with "Mumbai."

```python
#Question 8
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generate 1000 samples from a normal distribution
np.random.seed(42) # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)
df = pd.DataFrame(data, columns=['Value'])

# 2. Introduce 50 random missing values (NaN)
indices_to_replace = np.random.choice(df.index, size=50, replace=False)
df_missing = df.copy()
df_missing.loc[indices_to_replace, 'Value'] = np.nan

# 3. Fill missing values with the column mean
# We calculate the mean of the available data and apply it to the gaps
mean_val = df_missing['Value'].mean()
df_imputed = df_missing.fillna(mean_val)

# 4. Plot histograms before and after imputation
plt.figure(figsize=(12, 6))

# Histogram Before Imputation (with missing values)
plt.subplot(1, 2, 1)
plt.hist(df_missing['Value'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title(f'Before Imputation\n(Missing: 50, Mean: {mean_val:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram After Imputation
plt.subplot(1, 2, 2)
plt.hist(df_imputed['Value'], bins=30, color='salmon', edgecolor='black')
plt.title(f'After Imputation\n(Filled with {mean_val:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Imputation: If the missing values are less than 20–30%, I would use Median Imputation. I prefer the median over the mean because age can sometimes be skewed, and the median is more representative of the typical customer.

Segmented Imputation: A more advanced approach I might use is to fill missing ages based on "Payment Method"—for example, using the median age of "Credit Card" users to fill gaps for missing credit card transactions.

Step 3: Managing Outliers (Transaction Amount)
In retail, transaction amounts often have massive outliers (e.g., a bulk corporate purchase vs. a single candy bar).

Winsorization: I would likely use this to "cap" the extreme transactions at the 99th percentile. This keeps the high-value data in the model but prevents a single $50,000 transaction from distorting the entire learning process.

Log Transformation: If the transaction amounts are heavily right-skewed, I might apply a Log Transform to compress the range and make the distribution more normal.

Step 4: Categorical Encoding (Payment Method)

Stratified Splitting: When I split my data into Training and Testing sets, I must ensure both sets have the exact same ratio of fraud to non-fraud.
