###Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Descriptive statistics and inferential statistics are two main branches of statistical analysis. They serve different goals and use different methods. Here’s a clear comparison with examples.

### Overview

*   **Descriptive statistics** summarize and describe the features of a dataset you have. They tell you what the data looks like.
*   **Inferential statistics** use sample data to make conclusions, predictions, or generalizations about a larger population from which the sample was drawn, and they quantify uncertainty about those conclusions.

### Key differences

1.  **Purpose**

    *   **Descriptive:** Summarize observed data (what is).
    *   **Inferential:** Make decisions or predictions about a population based on a sample (what might be).

2.  **Data scope**

    *   **Descriptive:** Uses the full dataset at hand (a sample or population) and does not generalize beyond it.
    *   **Inferential:** Uses a sample to draw conclusions about a larger population.

3.  **Typical outputs**

    *   **Descriptive:** Measures of central tendency (mean, median, mode), measures of spread (range, variance, standard deviation, IQR), frequency counts, tables, charts (histograms, boxplots).
    *   **Inferential:** Point estimates, confidence intervals, hypothesis tests (p-values), regression coefficients with standard errors, predictions, Bayesian posterior distributions.

4.  **Role of probability**

    *   **Descriptive:** Little or no direct use of probability theory; just calculations on observed values.
    *   **Inferential:** Heavily uses probability to model sampling variability and to quantify uncertainty.

### Examples

**Example 1 — Average height in a classroom**

*   **Descriptive:** You measure heights of all 30 students in a classroom and compute the mean = 165 cm, median = 164 cm, standard deviation = 7 cm. These describe the classroom’s heights.
*   **Inferential:** You measure heights of 30 randomly chosen people from a city (a sample) and compute a sample mean = 170 cm with a 95% confidence interval 168–172 cm. You use that interval to infer the city’s average height and report uncertainty. You might also test whether the city mean differs from a known national mean using a t-test.

**Example 2 — Customer satisfaction survey**

*   **Descriptive:** A company surveys 200 customers and reports that 72% are satisfied, lists the distribution of ratings, and displays a bar chart. These describe responses in that sample.
*   **Inferential:** The company uses the 200 responses to estimate the satisfaction rate for its entire customer base (e.g., 72% ± 5% at 95% confidence) and tests whether a new policy changed satisfaction compared with last year (hypothesis test).

**Example 3 — Drug trial**

*   **Descriptive:** In a clinical trial, researchers report the number of participants experiencing side effects in each treatment arm, plus mean blood-pressure change and standard deviations.
*   **Inferential:** Researchers perform hypothesis tests or compute confidence intervals to decide whether the drug’s effect is statistically significant versus placebo, and estimate the size of the effect in the broader patient population.

### When to use each

*   Use descriptive statistics whenever you need to summarize, visualize, or communicate the characteristics of the data you have.
*   Use inferential statistics when you need to generalize beyond your sample, estimate population parameters, test hypotheses, or make predictions.

### Practical note about interpretation

*   A descriptive statement (e.g., “sample mean = 50”) is a fact about the data measured.
*   An inferential statement (e.g., “we are 95% confident the population mean is between 48 and 52”) includes uncertainty — it relies on assumptions (random sampling, model correctness, sample size adequacy). Always check those assumptions before trusting inferential results.

### Short summary

*   **Descriptive** = summarize observed data.
*   **Inferential** = draw conclusions about a larger population from a sample, with quantified uncertainty.

###Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Sampling is the process of selecting a subset (sample) from a larger group (population) to learn about the population without measuring everyone. Good sampling aims to produce a sample that accurately represents the population so that conclusions drawn from the sample (estimates, tests, predictions) are valid. Sampling is used because measuring the entire population is often impractical, expensive, or impossible.

### Key terms

*   **Population:** the full set of items or individuals you want to study (e.g., all voters in a country).
*   **Sample:** the subset you actually observe (e.g., 1,200 voters surveyed).
*   **Sampling frame:** the list or mechanism from which you draw the sample (e.g., voter registry).
*   **Sampling error:** the difference between sample estimates and true population values due to only observing a subset.
*   **Bias:** systematic error introduced by the sampling method that makes the sample unrepresentative.

### Random sampling vs. stratified sampling

Both are probability sampling methods (every unit has a known probability of selection), but they differ in how units are chosen.

*   **Definition**
    *   **Random sampling (simple random sampling, SRS):** every possible subset of the population of a given size has an equal chance of being selected. Practically, each individual is chosen independently with equal probability.
    *   **Stratified sampling:** the population is divided into distinct subgroups (strata) that are internally similar on some characteristic(s), then random samples are drawn from each stratum (often proportionally to stratum size, but sometimes with other allocation rules).
*   **Purpose**
    *   **Random sampling:** ensures overall fairness and simplicity; useful when the population is fairly homogeneous or when no strong subgroup effects are expected.
    *   **Stratified sampling:** ensures adequate representation of key subgroups (e.g., age groups, regions) and usually reduces sampling variance when strata differ from each other.
*   **When to use**
    *   **Random sampling:** when you can draw from the whole population easily, and you don't need guaranteed representation of small subgroups.
    *   **Stratified sampling:** when the population contains distinct groups with different characteristics relevant to the study, or when you want more precise estimates for those groups (or for the overall population).
*   **Advantages**
    *   **Random sampling:**
        *   Simple and easy to implement (with a good sampling frame).
        *   Unbiased estimator properties are straightforward.
    *   **Stratified sampling:**
        *   Improves precision (lower variance) of estimates if strata are homogeneous internally and heterogeneous between each other.
        *   Ensures small but important subgroups are included (avoids their underrepresentation).
        *   Allows subgroup-specific estimates (e.g., separate means for men and women).
*   **Disadvantages / challenges**
    *   **Random sampling:**
        *   Can by chance underrepresent small subgroups.
        *   Requires a complete, accurate sampling frame.
    *   **Stratified sampling:**
        *   Requires knowledge of relevant stratifying variables and a sampling frame that indicates stratum membership.
        *   More complex to design and analyze (weights may be needed if sampling fractions differ).

### Example

Example population: 10,000 registered voters in a city, proportions: 60% urban, 40% suburban.

*   **Simple random sampling (SRS):**
    *   Choose 400 voters at random from the full list. By chance you might get 220 urban and 180 suburban, or 260 urban and 140 suburban. Estimates for the overall city are unbiased, but subgroup representation can vary.
*   **Stratified sampling:**
    *   Split the list into two strata: urban (6,000) and suburban (4,000).
    *   Draw 240 urban and 160 suburban respondents (proportional allocation for total n = 400). This guarantees proportional representation and typically yields more precise estimates of the citywide vote share than SRS, especially if voting patterns differ between urban and suburban voters.
    *   Alternatively, you might oversample suburban voters (e.g., 200 suburban, 200 urban) to get better estimates for the smaller suburban group; then apply sampling weights in analysis to produce correct overall estimates.

### Practical notes

*   **Analysis:** Stratified samples often require using stratum weights or accounting for design in variance estimation (e.g., using survey-weighted methods) if sampling fractions differ across strata.
*   **Combine approaches:** Many surveys use multistage designs (cluster sampling + stratification) to balance cost and precision.
*   Nonresponse and frame errors can introduce bias regardless of sampling method; good survey practice includes measures to reduce and adjust for these.

### Short summary

*   **Sampling** picks a subset to learn about a population.
*   **Random (simple) sampling** gives every individual equal chance; it’s simple but can underrepresent small groups by chance.
*   **Stratified sampling** divides the population into subgroups and samples within them to guarantee representation and usually increase precision when groups differ.



# Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

### Definitions

*   **Mean (arithmetic mean):** The mean is the sum of all observations divided by the number of observations.
    If the dataset is $x_1, x_2, \ldots, x_n$, the mean is $\bar{x} = \frac{1}{n}\sum_{i=1}^n x_i$.

*   **Median:** The median is the middle value when the observations are ordered from smallest to largest.

    *   If $n$ is odd, the median is the $(n+1)/2$-th ordered value.
    *   If $n$ is even, the median is the average of the two middle ordered values (the $n/2$-th and $(n/2)+1$-th).

*   **Mode:** The mode is the value(s) that occur most frequently in the dataset. A dataset can have one mode (unimodal), more than one mode (multimodal), or no distinct mode if all values occur with equal frequency.

### Why these measures matter

*   **Summarize a dataset with a single, interpretable value:**

    These statistics provide a quick sense of the "center" or typical value in the data, which helps communicate and compare datasets efficiently.
*   **Different robustness to outliers and skew:**

    *   Mean is sensitive to every value, so large outliers or skewed distributions shift the mean markedly. This is useful when you want to account for all values (e.g., total/average cost), but can be misleading for skewed data.
    *   Median is robust to extreme values and better represents the typical observation for skewed distributions (e.g., household income).
    *   Mode highlights the most common outcome and is especially useful for categorical data (e.g., most frequent shoe size, most chosen survey option).
*   **Different uses and interpretations:**

    *   Mean is the natural choice for many mathematical procedures (e.g., in regression, variance calculation, and when combining data) and for estimating population averages.
    *   Median is preferred when you want a middle point resistant to outliers or when the distribution is not symmetric.
    *   Mode is important when the most frequent category is of interest (e.g., modal class in a histogram, most common defect type).
*   **Basis for further analysis:**

    *   The mean is used in variance and standard deviation calculations, hypothesis tests, confidence intervals, and many inferential procedures.
    *   The median ties into nonparametric statistics (e.g., Mann–Whitney test) and measures of skewness.
    *   Knowing the mode(s) helps identify multimodality, which may indicate distinct subgroups or different processes in the data.

### Examples (brief)

Dataset: {2, 3, 4, 8, 100}

*   Mean = (2+3+4+8+100)/5 = 23.4 (strongly affected by the outlier 100)
*   Median = 4 (better represents the typical value)
*   Mode = none (all values unique)

Categorical example (colors chosen): {red, blue, blue, green, blue, red}

*   Mode = blue (most frequent category)
*   Mean/median are not applicable for nominal categories.

### Practical guidance

*   For symmetric, outlier-free numeric data, mean and median will be similar; mean is often preferred for efficiency and inferential uses.
*   For skewed numeric data, report the median (and an interquartile range) to summarize central tendency robustly.
*   For categorical or discrete data, report the mode (and proportions) to show the most common category.
*   Often report more than one measure (e.g., mean and median) plus a measure of spread (standard deviation or IQR) to give a fuller picture of the data.

# Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

### Skewness

Skewness measures the asymmetry of a distribution around its mean.

*   Informal: It tells you whether the tail on one side of the distribution is longer or fatter than the other.
*   Common numeric measure (sample skewness):
    $g_1 = \frac{\frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})^3}{(\frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})^2)^{3/2}}$

    (There are slight variations in how software reports skewness; some use bias adjustments.)

### Interpretation

*   Skewness $\approx 0$: distribution roughly symmetric.
*   Positive skew (right skew): long or heavy tail to the right (toward larger values).
*   Negative skew (left skew): long or heavy tail to the left (toward smaller values).

### What a positive skew implies

*   The bulk (mass) of the data lies to the left of the mean; a few relatively large values pull the mean to the right.
*   Mean > median (typically) > mode (for many common shapes).
*   Examples: household income, waiting times, and many biological measures that have a lower bound at zero and a long right tail.
*   Practical consequence: the mean is influenced by extreme high values; median is often a better "typical" measure for positively skewed data.

### Kurtosis

Kurtosis measures the "tailedness" or peakedness of a distribution relative to the normal distribution. The most common (excess) kurtosis is:

*   excess kurtosis $= g_2 = \frac{\frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})^4}{(\frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})^2)^2} - 3$

    (the minus 3 makes the normal distribution have excess kurtosis = 0).

### Interpretation

*   Excess kurtosis = 0: tails similar to normal (mesokurtic).
*   Excess kurtosis > 0: heavier tails and often a sharper peak than normal (leptokurtic) — more frequent extreme values/outliers.
*   Excess kurtosis < 0: lighter tails and a flatter peak than normal (platykurtic) — fewer extremes.

### Notes and cautions

*   Kurtosis focuses on tails more than "peak"; two distributions can have the same kurtosis but look different in the center.
*   High kurtosis signals greater probability of outliers (extreme deviations) compared to a normal distribution.

### Quick summary

*   **Skewness** = asymmetry. Positive skew = right tail longer; mean pulled right; median < mean.
*   **Kurtosis** = tail heaviness/propensity for outliers. Positive (excess) kurtosis = heavier tails than normal, more extremes.

```python
#Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers. numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

from collections import Counter
from typing import List, Tuple

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

def mean(data: List[float]) -> float:
    if not data:
        raise ValueError("mean requires at least one data point")
    return sum(data) / len(data)

def median(data: List[float]) -> float:
    n = len(data)
    if n == 0:
        raise ValueError("median requires at least one data point")
    sorted_data = sorted(data)
    mid = n // 2
    if n % 2 == 1:
        return sorted_data[mid]
    else:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2

def mode(data: List[float]) -> List[float]:
    if not data:
        raise ValueError("mode requires at least one data point")
    counts = Counter(data)
    max_count = max(counts.values())
    # Return all values that have the highest frequency (handles multimodal)
    modes = [val for val, cnt in counts.items() if cnt == max_count]
    return sorted(modes)

# Compute and print results
print("Data:", numbers)
print("Mean:", mean(numbers))
print("Median:", median(numbers))
print("Mode(s):", mode(numbers))
```

### Python code to compute these exactly:

```python
#Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60] Given: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60]
import math

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

def mean(data):
    return sum(data) / len(data)

def sample_covariance(x, y):
    n = len(x)
    if n != len(y):
        raise ValueError("Lists must have same length")
    mx = mean(x)
    my = mean(y)
    cov_sum = sum((xi - mx) * (yi - my) for xi, yi in zip(x, y))
    return cov_sum / (n - 1)   # sample covariance

def sample_std(data):
    n = len(data)
    m = mean(data)
    var_sum = sum((xi - m) ** 2 for xi in data)
    sample_var = var_sum / (n - 1)
    return math.sqrt(sample_var)

cov_xy = sample_covariance(list_x, list_y)
sd_x = sample_std(list_x)
sd_y = sample_std(list_y)
r = cov_xy / (sd_x * sd_y)

print("Sample covariance:", cov_xy)
print("Sample standard deviation x:", sd_x)
print("Sample standard deviation y:", sd_y)
print("Pearson correlation r:", r)
```

```python
#Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
'''Explanation of results (for this dataset)

Five-number summary (computed by the script):
Min = 12
Q1 = 15.25
Median = 21.5
Q3 = 24.75
Max = 35
IQR = Q3 − Q1 = 9.5
Outlier fences:
Lower fence = Q1 − 1.5·IQR = 15.25 − 14.25 = 1.0
Upper fence = Q3 + 1.5·IQR = 24.75 + 14.25 = 39.0
Outliers by the 1.5·IQR rule: none, because all observations lie between 1.0 and 39.0.
The largest value 35 is high relative to the rest but does not exceed the upper fence (39.0), so it is not flagged as an outlier by this rule.
Interpretation

The distribution is roughly concentrated between ~14 and ~29, with a median of 21.5.
The middle 50% (Q1 to Q3) spans about 15.25 to 24.75, so the central spread (IQR) is 9.5.
There is some right skew (the highest value, 35, stretches the upper range more than the lower range), but not extreme enough to be an outlier under the standard 1.5·IQR criterion.
If you wanted to flag more extreme values only, you could use the 3·IQR rule (outer fences) which would be even wider; conversely, for stricter detection, lower the multiplier (<1.5) but that is nonstandard.
Notes

The script uses numpy percentile with interpolation='midpoint' which is appropriate for small samples; in numpy >=1.22 the kwarg is method='midpoint' instead—if you see a deprecation warning, replace interpolation='midpoint' with method='midpoint'.
You can change flierprops or markersize to style the outliers display.'''
import numpy as np
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Convert to numpy array for convenience
arr = np.array(data)

# Five-number summary
minimum = np.min(arr)
q1 = np.percentile(arr, 25, interpolation='midpoint')  # numpy >=1.22: use method='midpoint'
median = np.median(arr)
q3 = np.percentile(arr, 75, interpolation='midpoint')
maximum = np.max(arr)

# IQR and outlier fences
iqr = q3 - q1
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

# Identify outliers
outliers = arr[(arr < lower_fence) | (arr > upper_fence)]

print("Five-number summary:")
print(f" Min  = {minimum}")
print(f" Q1   = {q1}")
print(f" Median = {median}")
print(f" Q3   = {q3}")
print(f" Max  = {maximum}")
print()
print(f"IQR = Q3 - Q1 = {iqr}")
print(f"Lower fence = Q1 - 1.5*IQR = {lower_fence}")
print(f"Upper fence = Q3 + 1.5*IQR = {upper_fence}")
print()
print("Outliers (by 1.5*IQR rule):", outliers.tolist())

# Create a boxplot
fig, ax = plt.subplots(figsize=(6, 3))
ax.boxplot(arr, vert=False, showmeans=True, meanline=True,
           flierprops=dict(marker='o', color='red', markersize=6))
ax.set_yticks([1])
ax.set_yticklabels(['data'])
ax.set_title('Boxplot (data)')
ax.grid(axis='x', linestyle='--', alpha=0.4)

# Annotate outliers on the plot
for out in outliers:
    ax.annotate(str(out), xy=(out, 1.05), xytext=(0, 8),
                textcoords='offset points', ha='center', fontsize=9, color='red')

plt.show()
```

```python
'''Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. ● Explain how you would use covariance and correlation to explore this relationship. ● Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500]   daily_sales = [2200, 2450, 2750, 3200, 4000]

How to use covariance and correlation to explore the relationship
Covariance:

Measures the joint variability of advertising spend and daily sales.
A positive covariance means that when ad spend increases, sales tend to increase too; a negative covariance means they tend to move in opposite directions.
Magnitude depends on units (dollars×dollars) so it’s not directly comparable across different variable pairs.
Correlation (Pearson’s r):

Standardizes covariance by the product of the variables’ standard deviations, giving a dimensionless value in [−1, +1].
r close to +1 indicates a strong positive linear relationship, r close to −1 a strong negative linear relationship, and r near 0 little or no linear relationship.
Use correlation to quantify strength and direction of linear association and to compare relationships across variable pairs.
Practical workflow

Step 1: Plot the data (scatter plot) to inspect linearity, outliers, and any nonlinearity.
Step 2: Compute covariance to confirm direction of association.
Step 3: Compute Pearson correlation to quantify strength.
Step 4 (optional): Fit a simple linear regression (sales ~ ad spend) to estimate effect size and test significance.
Remember: correlation ≠ causation. Investigate confounders and use experiments (A/B tests) for causal claims.'''

import math
import matplotlib.pyplot as plt

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

def mean(xs):
    return sum(xs) / len(xs)

def sample_covariance(x, y):
    n = len(x)
    mx, my = mean(x), mean(y)
    return sum((xi - mx) * (yi - my) for xi, yi in zip(x, y)) / (n - 1)

def sample_std(xs):
    n = len(xs)
    m = mean(xs)
    return math.sqrt(sum((xi - m) ** 2 for xi in xs) / (n - 1))

# Compute metrics
cov_xy = sample_covariance(advertising_spend, daily_sales)
sd_x = sample_std(advertising_spend)
sd_y = sample_std(daily_sales)
pearson_r = cov_xy / (sd_x * sd_y)

# Print results
print(f"Sample covariance (ad_spend, daily_sales): {cov_xy:.1f}")
print(f"Std dev ad_spend: {sd_x:.4f}")
print(f"Std dev daily_sales: {sd_y:.4f}")
print(f"Pearson correlation coefficient r: {pearson_r:.4f}")

# Optional: quick scatter plot with least-squares fit line
n = len(advertising_spend)
mx, my = mean(advertising_spend), mean(daily_sales)
var_x = sum((xi - mx)**2 for xi in advertising_spend) / (n - 1)
slope = cov_xy / var_x
intercept = my - slope * mx

plt.scatter(advertising_spend, daily_sales, label='Data')
x_vals = [min(advertising_spend), max(advertising_spend)]
y_vals = [intercept + slope * x for x in x_vals]
plt.plot(x_vals, y_vals, color='orange', label=f'Fit: y = {slope:.2f}x + {intercept:.1f}')
plt.xlabel('Advertising spend')
plt.ylabel('Daily sales')
plt.title('Advertising spend vs Daily sales')
plt.legend()
plt.grid(True)
plt.show()

'''
Results (from this code)
Sample covariance ≈ 20,375.0
Pearson correlation coefficient r ≈ 0.9942
Interpretation

Covariance > 0: ad spend and sales move together (higher spend associated with higher sales).
r ≈ 0.9942: extremely strong positive linear relationship in this sample — advertising spend explains most of the linear variation in daily sales here.
Caveats: small sample (n=5) and observational data. Correlation alone doesn’t prove causation; consider plotting residuals, checking for outliers or nonlinearity, and using controlled experiments to test causality.'''
```

```python
#Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product. ● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use. ● Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
'''Summary statistics:

Mean: The average score, which gives a central value (7.33 for this data).

Standard deviation: Measures the spread or variability of the scores (1.58 for this data).

Median: The middle value when scores are ordered, showing the typical score (7 for this data).

Minimum and maximum: The lowest and highest scores (4 and 10 for this data).

Visualizations:

Histogram: To visualize the frequency distribution of scores and identify patterns like skewness or modality.

Box plot (optional): To detect outliers and understand the interquartile range.'''

import matplotlib.pyplot as plt
import numpy as np

# Customer satisfaction survey data (Scale 1-10)
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# --- 1. Calculate Summary Statistics ---
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
std_dev = np.std(survey_scores)

print(f"Summary Statistics:")
print(f"  Mean Score: {mean_score:.2f}")
print(f"  Median Score: {median_score}")
print(f"  Standard Deviation: {std_dev:.2f}")

# --- 2. Create Histogram Visualization ---

# Determine bin edges to properly visualize discrete 1-10 scores.
# Bins are set to center on the integers (4, 5, 6, 7, 8, 9, 10).
# Edges go from 3.5 up to (but not including) 11.5, in steps of 1.
bins = np.arange(3.5, 11.5, 1)

plt.figure(figsize=(10, 6))

# Plot the histogram
# rwidth=0.8 adds spacing between the bars for better readability
plt.hist(survey_scores, bins=bins, edgecolor='black', alpha=0.85, color='#4c72b0', rwidth=0.8)

# Set the x-axis ticks to show the discrete survey scores
unique_scores = sorted(list(set(survey_scores)))
plt.xticks(unique_scores, fontsize=12)

# Add title and labels
plt.title('Distribution of Customer Satisfaction Survey Scores (Scale 1-10)', fontsize=16, fontweight='bold')
plt.xlabel('Survey Score (1-10)', fontsize=14)
plt.ylabel('Frequency (Count)', fontsize=14)

# Add a dashed line for the Mean score
plt.axvline(mean_score, color='red', linestyle='--', linewidth=1.5, label=f'Mean: {mean_score:.2f}')

# Ensure y-axis shows integer counts
max_freq = max([c.get_height() for c in plt.gca().patches])
plt.yticks(np.arange(0, max_freq + 1, 1))

plt.legend(loc='upper right')
plt.grid(axis='y', alpha=0.4)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

print("\nHistogram plot generated successfully.")
```
