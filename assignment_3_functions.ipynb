# Assignment 3 Functions

1. Difference between a function and a method in Python     
Function: A block of reusable code which is independent and can be called by name.
Method: A function that is associated with an object and called on that object (e.g., an instance of a class).

2. Concept of function arguments and parameters in Python   
Parameters are the variables listed in the function definition.
Arguments are the actual values passed to a function when it is called.

3. Different ways to define and call a function in Python   
Define:     
Using def keyword
Using lambda (anonymous functions)  
Call:   
By function name and parentheses with arguments

4. Purpose of the return statement in a Python function     
The return statement sends a value back to the caller and exits the function.

5. Iterators vs Iterables in Python     
Iterable: An object that can return an iterator (e.g., list, tuple, string).
Iterator: An object representing a stream of data; has a __next__() method to fetch next item.

6. Generators in Python and how they are defined    
Generators are special functions that yield values one at a time using yield keyword, allowing lazy evaluation.

7. Advantages of using generators over regular functions    
Memory efficient (do not store entire sequence)
Lazily evaluated (generate values on-the-fly)
Can represent infinite sequences

8. Lambda function in Python and when is it used    
A small anonymous function defined with lambda keyword, useful for short, simple functions usually as arguments to higher-order functions.

9. Purpose and usage of map() function in Python    
map() applies a function to all items in an iterable.

10. Difference between map(), reduce(), and filter()    
map(func, iterable) — applies func to every item.
filter(func, iterable) — selects items where func returns True.
reduce(func, iterable) — applies func cumulatively to reduce the iterable to a single value (needs from functools import reduce)

```python
# 1. Sum of all even numbers in a list:

def sum_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)

print(sum_even([1, 2, 3, 4, 5, 6]))
```

```python
# 2. Reverse a string:

def reverse_string(s):
    return s[::-1]

print(reverse_string("Python"))
```

```python
# 3. Return squares of numbers in a list:

def square_list(numbers):
    return [num ** 2 for num in numbers]

print(square_list([1, 2, 3]))
```

```python
# 4. Check if a number is prime (1-200):

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

primes = [x for x in range(1, 201) if is_prime(x)]
print(primes)
```

```python
#5. Iterator class to generate Fibonacci sequence:

class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return val

fib_iter = Fibonacci(10)
for num in fib_iter:
    print(num, end=" ")
```

```python
#6. Generator for powers of 2 up to given exponent:

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power, end=" ")
```

```python
#7. Generator to read a file line by line:

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
# 8. Lambda to sort list of tuples by second element:

data = [(1, 3), (4, 1), (5, 2)]
data_sorted = sorted(data, key=lambda x: x[1])
print(data_sorted)  # Output: [(4, 1), (5, 2), (1, 3)]
```

```python
# 9. Program using map() to convert Celsius to Fahrenheit:

def c_to_f(c):
    return (c * 9/5) + 32

temps_c = [0, 10, 20, 30]
temps_f = list(map(c_to_f, temps_c))
print(temps_f)
```

```python
#10. Program using filter() to remove vowels from a string:

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda x: x not in vowels, s))

print(remove_vowels("Hello World"))
```

```python
#11. Book shop orders with lambda and map:
orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99)
]

def order_total(order):
    order_num, _, quantity, price = order
    product = quantity * price
    if product < 100:
        product += 10
    return (order_num, round(product, 2))

totals = list(map(lambda order: order_total(order), orders))
print(totals)
```
