**Question 3**: What is heteroscedasticity, and why is it important to address in regression models?

**Answer-** Heteroscedasticity occurs when the variance of the residuals (errors) in a regression model is not constant across all levels of the independent variable.1 In simpler terms, the "spread" or "scatter" of the data points around the regression line changes as the value of the predictor variable changes.Visualizing HeteroscedasticityIn a Residual vs. Fitted plot, heteroscedasticity often appears as a "fan" or "cone" shape.2 This indicates that the model's predictions are more accurate for some ranges of data and much less accurate for others.

Why it is Important to Address:

Addressing heteroscedasticity is critical for the following reasons:Unreliable Statistical Tests:Heteroscedasticity invalidates the standard errors of the regression coefficients. Since $t$-tests and $F$-tests rely on these standard errors, your $p$-values may be misleading. You might conclude that a variable is "significant" when it actually isn't (Type I error).Invalid Confidence Intervals:Because the standard errors are biased, the confidence intervals for your predictions will be either too narrow or too wide, leading to incorrect conclusions about the precision of your model.Loss of Efficiency:While the OLS (Ordinary Least Squares) estimates of the coefficients remain unbiased even with heteroscedasticity, they are no longer the Best Linear Unbiased Estimators (BLUE). This means there is likely a better way to fit the model that would result in smaller errors.How to Detect and Fix ItDetection:Visual: Look for fan-shaped patterns in a Residuals vs. Fitted plot.Statistical Tests: Perform the Breusch-Pagan test or the White test.Common Fixes:Log Transformation: Apply a natural log to the dependent variable (3$Y$) to stabilize the variance.4Weighted Least Squares (WLS): Use a different regression method that gives less weight to observations with higher variance.Robust Standard Errors: Use "Heteroscedasticity-Consistent" standard errors (like Huber-White standard errors) to fix the $p$-values without changing the coefficients.

**Question 6:**  Implement a Python program to fit a Simple Linear Regression model to the following sample data: ● X = [1, 2, 3, 4, 5] ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Define the sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) # Reshape to 2D array for the model
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# 2. Initialize and fit the Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# 3. Calculate predictions for the regression line
Y_pred = model.predict(X)

# 4. Plotting the results
plt.scatter(X, Y, color='blue', label='Actual Data Points') # Plot original data
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line') # Plot regression line
plt.title('Simple Linear Regression')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.legend()
plt.grid(True)
plt.show()

# Optional: Print the model coefficients
print(f"Slope (m): {model.coef_[0]:.2f}")
print(f"Intercept (c): {model.intercept_:.2f}")
```

**Question 7**: Fit a Multiple Linear Regression model on this sample data: ● Area = [1200, 1500, 1800, 2000] ● Rooms = [2, 3, 3, 4] ● Price = [250000, 300000, 320000, 370000] Check for multicollinearity using VIF and report the results.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# 1. Define the sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# 2. Fit the Multiple Linear Regression model
X = df[['Area', 'Rooms']]
Y = df['Price']

model = LinearRegression()
model.fit(X, Y)

# 3. Calculate VIF for multicollinearity
# VIF requires adding a constant (intercept) to the features
X_vif = add_constant(X)
vif_data = pd.DataFrame()
vif_data["Feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]

# 4. Report Results
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
print("\nVIF Results:")
print(vif_data[1:]) # Excluding the constant for readability
```

**Question 8**: Implement polynomial regression on the following data: ● X = [1, 2, 3, 4, 5] 3 ● Y = [2.2, 4.8, 7.5, 11.2, 14.7] Fit a 2nd-degree polynomial and plot the resulting curve.


The 2nd-degree polynomial regression model for the given data results in the following estimated equation:\(Y=0.06+1.94X+0.20X^{2}\)This model captures the non-linear trend by mapping the original feature \(X\) to a higher-dimensional space including \(X^{2}\), allowing for a curved line of best fit.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 1. Define the data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# 2. Transform features to degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# 3. Fit the model
model = LinearRegression()
model.fit(X_poly, Y)

# 4. Generate points for a smooth curve
X_smooth = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_smooth_poly = poly.transform(X_smooth)
Y_smooth = model.predict(X_smooth_poly)

# 5. Plotting
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_smooth, Y_smooth, color='red', label='2nd-Degree Polynomial Fit')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Print Coefficients
print(f"Equation: Y = {model.intercept_:.2f} + {model.coef_[1]:.2f}X + {model.coef_[2]:.2f}X^2")
```

**Question 10**: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

**Answer**-

1. Address Multicollinearity
Multicollinearity occurs when independent variables (e.g., area and number of rooms) are highly correlated, making coefficient estimates unstable.
Variable Selection (VIF-based): Calculate the Variance Inflation Factor (VIF) for each feature. If a feature has a VIF greater than 5 or 10, consider removing it or combining it with another to reduce redundancy.
Feature Engineering: Combine highly correlated features into a single index. For example, instead of using "Number of Rooms" and "Square Footage" separately, create a "Total Livable Space" metric.
Regularization (Ridge/Lasso Regression): Apply Lasso (L1) or Ridge (L2) regression. These techniques introduce a penalty for large coefficients, which helps manage correlated predictors and prevents overfitting.
Principal Component Analysis (PCA): Transform correlated features into a smaller set of uncorrelated components (Principal Components) to use as new predictors.


2. Address Heteroscedasticity
Heteroscedasticity occurs when the variance of the error terms is not constant across all levels of an independent variable (e.g., prices for expensive homes vary more than for cheap ones).
Logarithmic Transformation: Apply a log transformation to the dependent variable (Price). This "squashes" the data range and often stabilizes the variance, making the errors more homoscedastic.
Weighted Least Squares (WLS): Use WLS regression to assign higher weights to data points with lower variance and lower weights to points with higher variance, ensuring that outliers do not disproportionately influence the model.
Robust Standard Errors: Use Huber-White standard errors (sandwich estimators) during the estimation process. This doesn't change the coefficients but provides more accurate p-values and confidence intervals even if heteroscedasticity remains.
Sub-market Analysis: Define and model separate sub-markets (e.g., by city or neighborhood). This accounts for different price behaviors in various locations that might cause uneven variance when pooled together.


3. Ensure Model Robustness
Cross-Validation: Use k-fold cross-validation to ensure the model generalizes well to unseen data rather than just fitting the training noise.
Ensemble Methods: In 2026, data scientists frequently use XGBoost or Stacked Models (combining multiple models) to capture complex non-linear relationships that simple linear regression might miss, often naturally handling some variance issues.
