# Therotical questions

1. What is NumPy, and why is it widely used in Python?  
NumPy (Numerical Python) is a library for numerical computations in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently. It is widely used due to its speed, efficiency, and convenience for numerical operations compared to native Python lists.

2. How does broadcasting work in NumPy?   
Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by "stretching" the smaller array across the larger one without copying data, enabling element-wise operations in a memory-efficient way.

3. What is a Pandas DataFrame?  
A Pandas DataFrame is a two-dimensional, size-mutable, tabular data structure with labeled axes (rows and columns). It is akin to a spreadsheet or SQL table in Python, making data manipulation and analysis easy.

4. Explain the use of the groupby() method in Pandas.   
The groupby() method splits data into groups based on some criteria and allows you to perform aggregation, transformation, or filtration on each group independently.

5. Why is Seaborn preferred for statistical visualizations?     
Seaborn is built on Matplotlib and provides a higher-level interface for creating attractive, informative, and easy-to-understand statistical graphics with fewer lines of code and built-in support for complex statistical visualizations.

6. What are the differences between NumPy arrays and Python lists?    
NumPy arrays are homogeneous (all elements same type), support vectorized operations, are more memory-efficient, and faster for numerical operations, whereas Python lists are heterogeneous, slower in numerical computations, and more flexible for general use.

7. What is a heatmap, and when should it be used?   
A heatmap is a graphical representation of data where individual values are represented with colors. It is used to visualize correlation matrices or patterns in data intensity or distribution.

8. What does the term "vectorized operation" mean in NumPy?   
Vectorized operations refer to performing element-wise operations on entire arrays without explicit loops, which leads to optimized and faster execution.

9. How does Matplotlib differ from Plotly?    
Matplotlib is a static plotting library producing 2D visualizations primarily, while Plotly supports interactive, web-ready, and 3D plots with zooming, panning, and tooltips.

10. What is the significance of hierarchical indexing in Pandas?    
Hierarchical indexing (MultiIndex) allows multiple levels of indexing on an axis, enabling more complex data representation and easy access to subsets in higher-dimensional data.

11. What is the role of Seaborn's pairplot() function?    
pairplot() creates a matrix of scatterplots and histograms that help visualize pairwise relationships and distributions for multiple variables in a dataset.

12. What is the purpose of the describe() function in Pandas?   
describe() provides a summary of statistics such as count, mean, std, min, max, and quartiles for numerical columns in a DataFrame.

13. Why is handling missing data important in Pandas?   
Missing data can lead to incorrect analyses and errors in computations. Pandas provides tools to identify, remove, or impute missing values, ensuring data integrity.

14. What are the benefits of using Plotly for data visualization?   
Plotly offers interactive, high-quality, publication-ready plots that support zooming, hovering details, and can be embedded in web apps, enhancing data exploration.

15. How does NumPy handle multidimensional arrays?    
NumPy stores data in contiguous blocks of memory and uses shape and strides to efficiently index and manipulate multidimensional arrays.

16. What is the role of Bokeh in data visualization?    
Bokeh is a Python library for creating interactive, web-based visualizations that are easy to embed in web apps and dashboards.

17. Explain the difference between apply() and map() in Pandas.   
apply() is used on DataFrames or Series and allows applying any function to rows or columns, while map() is specifically used with Series to map values from a dictionary, function, or Series to corresponding elements.

18. What are some advanced features of NumPy?   
Advanced features include broadcasting, masked arrays, structured arrays, linear algebra support, Fourier transforms, and random number generation.

19. How does Pandas simplify time series analysis?      
Pandas offers datetime indexing, resampling, rolling windows, time shifting, and powerful date range generation that simplify time series manipulation and analysis.

20. What is the role of a pivot table in Pandas?    
Pivot tables summarize data by aggregating it when grouping by one or more keys, allowing quick reshaping and summarization of data tables.

21. Why is NumPy's array slicing faster than Python's list slicing?   
NumPy slicing returns views into the original array (no data copying), which makes it faster, while Python list slicing creates new copies of sublists.

22. What are some common use cases for Seaborn?   
Common use cases include visualizing distributions, relationships, categorical data, heatmaps, regression plots, and pairwise variable relationships.

```python
#1. How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = arr.sum(axis=1)
print(row_sums)
```

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
mean_A = df['A'].mean()
print(mean_A)
```

```python
#3. Create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.rand(5,5), columns=list('ABCDE'))
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#5. Generate a bar plot using Plotly.

import plotly.express as px
import pandas as pd

data = {'Category': ['A', 'B', 'C'], 'Values': [10, 15, 7]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Category', y='Values')
fig.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.

import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]})
df['B'] = df['A'] * 2
print(df)
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a * b
print(c)
```

```python
#8. Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 7]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd
df = pd.DataFrame({'A': [1, 5, 3, 9]})
filtered = df[df['A'] > 4]
print(filtered)
```

```python
#10. Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)
sns.histplot(data, kde=True)
plt.show()
```

```python
#11. Perform matrix multiplication using NumPy.

import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```
